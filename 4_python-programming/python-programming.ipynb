```python
list_plat_mobil = ['B 1234 AB', 'B 6721 TY', 'B 1233 AY', 'B 6629 DD', 'B 1111 AM', 'B 6726 D', 'D 11223 KJ', 'AE 44677 GH', 'AE 67269 AA']

list_ganjil = []
list_genap = []

for plat in list_plat_mobil:
    angka_plat = ''
    for huruf in plat:
        if huruf.isdigit():
            angka_plat += huruf

    angka_plat = int(angka_plat)
    if angka_plat % 2 == 0:
        list_genap.append(plat)
    else:
        list_ganjil.append(plat)

print("Jumlah plat genap :", len(list_genap))
print("Jumlah plat ganjil:", len(list_ganjil))
```

## soal 4

Buatlah sebuah function string_processor dengan inputan 2 parameter, yaitu suatu string 1 kata dan nama proses yang akan dilakukan. Function ini memiliki 4 proses
- first_char : return huruf pertama dari string yang diinput
- last_char : return huruf terakhir dari string yang diinput
- num_chars : return jumlah huruf string input
- num_vowels : return jumlah huruf vocal pada string input


```python
##tuliskode
def string_processor(input_string : 'str', process = 'first_char'):
  #membuang spasi pada awal dan akhir string
  input_string = input_string.strip()

  #process pada string input
  list_vowel = ['a', 'i', 'u', 'e', 'o']
  if process == 'first_char':
    result = input_string[0]
  elif process == 'last_char':
    result = input_string[-1]
  elif process == 'num_chars':
    list_huruf = [char for char in input_string if char != " "]
    result = len(list_huruf)
  elif process == 'num_vowels':
    list_vowel_in_input_string = [char for char in input_string if (char != " ") & (char in list_vowel)]
    result = len(list_vowel_in_input_string)

  #menampilkan hasil
  return result
```

## soal 6

Menggunakan dataset superstore! Buatlah function yang bisa menghitung Jumlah Sales dan Profit pada suatu tahun dan region tertentu dengan default tahun 2017. Hint : Buat kolom tahun

<b>Tentukan total sales dan profit:</b><br>
- South Region pada tahun 2016
- West Region pada tahun 2017

```python
#run baris kode ini lalu upload file superstore.xls disini dari lokal drive ke google collab ini
from google.colab import files
uploaded = files.upload()
```

```python
superstore = pd.read_excel('superstore.xls')

def hitung_sales_profit(data, region, tahun=2017):
    # Copy data
    data_used = data.copy()

    # Filter berdasarkan region
    condition_1 = (data_used['Region'] == region)
    data_used = data_used[condition_1]

    # Buat tambahan kolom tahun
    data_used['Order Year'] = data_used['Order Date'].dt.year

    # Filter berdasarkan tahun
    condition_2 = (data_used['Order Year'] == tahun)
    data_used = data_used[condition_2]

    # Hitung agregasi
    result = data_used.groupby(['Region', 'Order Year'], as_index=False).agg(
        total_sales=('Sales', 'sum'),
        total_profit=('Profit', 'sum'),
    )

    # Memberikan nilai kembalian saat memanggil fungsi
    return result

# Panggil fungsi hitung_sales_profit dengan kondisi yang diberikan
hasil_perhitungan = hitung_sales_profit(superstore, region='South', tahun=2016)
hasil_perhitungan2 = hitung_sales_profit(superstore, region='West', tahun=2017)

# Gabungkan hasil perhitungan dari kedua region
hasil_gabungan = pd.concat([hasil_perhitungan, hasil_perhitungan2], ignore_index=True)

print(hasil_gabungan)
```

```python
df_superstore = pd.read_excel('superstore.xls')
df_superstore.head(3)
```

step1: ekstraksi tahun

```python
#get year
df_superstore['year'] = df_superstore['Order Date'].dt.year #menggunakan fungsi .dt.year untuk mengambil data tahun dari kolom order_date
df_superstore['year'].tail (3)
```

step3: Agregasi

*   Agregasi dilakukan pada kolom yang paling penting yaitu sales, customer_id, customer_name dan tahun.
*   Agregasi dilakukan pada kolom2 tersebut agar mendapatkan nilai total penjualan(sales) per pelanggan untuk tiap tahun.


```python
df_superstore_group_year = df_superstore.groupby(['year','Customer ID', 'Customer Name'])['Sales'].sum().reset_index()
df_superstore_group_year.tail()
```

step 4: Statistik Deskriptif
 berdasarkan total penjualan pertahun yaitu dari data kolom year dan kolom sales

```python
df_superstore_group_year_desc = df_superstore_group_year.groupby(['year'])['Sales'].describe().reset_index()
df_superstore_group_year_desc
```

step 5:


*  Data asli hasil agregasi (hasil dari step3) dimerge/gabung dengan data statistik deskriptif total penjualan(hasil dari step 4), digabungkan berdasarkan kolom year dan disimpan dalam variable **df_superstore_group_year**
*  Dimana data statistik deskriptif  yang ditampilkan hanya 'year', 'mean', percentile percentile '25%', percentile '50%', percentile '75%'
*   Penggabungan ini dilakukan dengan metode inner, yang berarti hanya data yang memiliki tahun yang sama di kedua DataFrame yang akan digabung.




```python
df_superstore_group_year = df_superstore_group_year.merge(df_superstore_group_year_desc[['year', 'mean', '25%', '50%', '75%']], on='year', how='inner')
df_superstore_group_year.head()
```

Step 6: segmentasi customer

*   Segmentasi dibuat berdasarkan olah data statistik deskriptif dan hasil agregasi pada total penjualan pertahun yaitu data kolom sales, year
*   karena data total penjualan/sales adalah ukuran kinerja utama bidang usaha pertokoan (store)
*   segmentasi dibagi jadi 3 low, mid, high:
1.   low --> total sales per tahun pada percentile **25%**
2.   medium --> total sales per tahun pada percentile **50%**
3.   medium --> total sales per tahun pada percentile **75%**




```python
#function segmentation user besed on total sales yearly
def segment(x):
    if x['Sales'] < x['50%']:
        segment = 'Low'
    elif x['Sales'] <= x['75%'] and x['Sales'] >= x['50%']:
        segment = 'Mid'
    else:
        segment = 'High'
    return segment
```

Step 7: function Lambda untuk menambahkan kolom baru yaitu : segment, dimana kolom segment adalah informasi segmentasi total penjualan

```python
df_superstore_group_year['segment'] = df_superstore_group_year.apply(lambda x: segment(x), axis=1)
df_superstore_group_year.head()
```

Step 8: total jumlah customer pada masing-masing segment di tahun 2017

```python
#filter tahun 2017 dan count unique based on customer id
df_superstore_group_year[(df_superstore_group_year.year == 2017)].groupby(['segment'])['Customer ID'].nunique()
```
